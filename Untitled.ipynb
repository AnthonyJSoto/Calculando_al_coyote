### $Resolucion:$

Realizaremos el DCL del cuerpo cuadrado:<br>
![title](2.png)

$$gdl=6*1-4=2$$ <dt>
Realizando la sumatoria de fuerzas:
$$\Sigma F e_{r}=ma\vec{e_{r}}$$
$$N-mgcos(\theta)=m\left(-\frac{b}{2} \cdot w^{2} \right) \tag{1}$$<dt>
$$\Sigma F e_{\theta}=ma\vec{e_{\theta}}$$
$$-Fr+mgsen(\theta)=m\left(\frac{b}{2} \cdot \alpha \right) \tag{2}$$ <dt>
Se sabe que la sumatoria de torques es:
$$\Sigma \tau=I_{o} *\alpha \tag{3}$$ <dt> 
El momento de inercia aplicando Steiner en el cuadrado respecto de un eje perpendicular al plano de la figura que pasa por O es:
$$I_{o}=\frac{1}{12}m(b^{2}+b^{2})+m\left(\frac{b}{2}\right)^2$$
$$I_{o}=\frac{5}{12}mb^{2} \tag{4}$$<dt>
La sumatoria de torques respecto a O es:
$$\Sigma \tau_{O} =\frac{b}{2}mgsen(\theta) \tag{5}$$<dt>
Las ecuaciones 4 y 5 reemplazando en 3 se tiene:
$$\frac{b}{2}mgsen(\theta)=\frac{5}{12}mb^{2}*\alpha $$<br>
$$\alpha=\frac{6}{5} \frac{gsen(\theta)}{b} \tag{6}$$<dt>
Se tiene la aceleracion angular en la ecuación 5, entonces:
$$\alpha=\frac{dw}{dt}$$
$$\alpha d\theta=wdw $$
$$\frac{6}{5}*\frac{gsen(\theta)}{b} \cdot d\theta=w \cdot dw $$
$$\frac{6}{5} \cdot \frac{g}{b} \int_{0}^{\theta} sen(\theta) \cdot d\theta=\int_{0}^{w} w \cdot dw$$
$$-\frac{6}{5} \cdot \frac{g}{b} \cdot (cos(\theta)-1)=\frac{w^{2}}{2}$$
$$w^{2}=\frac{12}{5} \cdot \frac{g}{b} \cdot (1-cos{\theta}) $$
    
Utilizando la ecuacion 1:
$$N-mgcos(\theta)=m\left(-\frac{b}{2} \cdot w^{2} \right)$$
$$-N+mgcos(\theta)=m\left(\frac{6}{5} \cdot g \cdot (1-cos{\theta}) \right)$$
$$-N+mgcos(\theta)=\frac{6}{5} \cdot mg \cdot (1-cos\theta)$$ <br>
$$N=\frac{mg}{5} \cdot (11 \cdot cos\theta - 6)$$ <dt>
Utilizando la ecuacion 2:
$$-Fr+mgsen(\theta)=m\left(\frac{b}{2} \cdot \alpha \right)$$
$$-Fr+mgsen(\theta)=m\left(\frac{b}{2} \cdot \frac{6}{5} \frac{gsen(\theta)}{b} \right)$$
$$Fr= \frac{2}{5} mgsen\theta$$ <dt>
    
Las ecuaciones del movimientos son:
$$N=\frac{mg}{5} \cdot (11 \cdot cos\theta - 6) \tag{7}$$ <dt>
$$Fr= \frac{2}{5} mgsen\theta \tag{8}$$ <dt>

#### Respuesta literal a.

```python
#datos
m=2000
b=3
u=0.3
g=9.8
angulo=np.arange(0,57.00,0.01)
N=((m*g)/5)*(11*np.cos(angulo*(np.pi/180))-6)
fig=px.line(x=angulo,y=N,labels={'x':'Angulo de giro [Grados]','y':'Norma [N]'},title="Normal vs Angulo de giro ")
fig.show()
```

#### Respuesta literal b.

```python
#datos
m=2000
b=3
u=0.3
g=9.8
angulo=np.arange(0,57.00,0.01)
velang=np.sqrt((12*g*(1-np.cos(angulo*(np.pi/180))))/(5*b))
fig=px.line(x=angulo,y=velang,labels={'x':'Angulo de giro [Grados]','y':'Velocidad angular [grados/s]'},title="Velocidad angular vs Angulo de giro ")
fig.show()
```
